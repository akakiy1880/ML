# Задача 2

Сложные проценты - https://ru.onlinemschool.com/math/library/percent/percent10/

Пользователь желает положить деньги под проценты на определенный срок. Банк каждый год начисляет 8% годовых. Вычислить полученную сумму в конце срока.

Доп. задача, распишите прибыль за каждый год. (можно использовать циклы если знаете, но можно и без)

Онлайн калькулятор, для проверки ответов: https://wpcalc.com/slozhnyj-procent/

```python
# сумма пользователя в рублях, которую он положит в банк
source_amt = 412500
# годовой процент
annual_interest = 8

# срок в 5 лет
time = 5
time_final = 1
# в эту переменную нужно записать ответ
result = source_amt
```

```python
#Vesrion 1.0
result = source_amt * (1 + (annual_interest / 100))**time

profit1 = source_amt * (1 + (annual_interest / 100))**1 - source_amt
profit2 = source_amt * (1 + (annual_interest / 100))**2 - source_amt - profit1
profit3 = source_amt * (1 + (annual_interest / 100))**3 - source_amt - profit1 - profit2
profit4 = source_amt * (1 + (annual_interest / 100))**4 - source_amt - profit1 - profit2 - profit3
profit5 = source_amt * (1 + (annual_interest / 100))**5 - source_amt - profit1 - profit2 - profit3 - profit4
profit = profit1 + profit2 + profit3 + profit4 + profit5


print("\n                                            #Version 1.0\n")
print(f"Полученная сумма: {round(result,2)} руб. Общая прибыль: {round(profit,2)} руб")
print('\n-------------------------------------------------------------------------------\n')
print(f"Прибыль за 1-ый период: { round(profit1,2) } руб")
print(f"Прибыль за 2-ый период: { round(profit2,2) } руб")
print(f"Прибыль за 3-ый период: { round(profit3,2) } руб")
print(f"Прибыль за 4-ый период: { round(profit4,2) } руб")
print(f"Прибыль за 5-ый период: { round(profit5,2) } руб")
print('\n-------------------------------------------------------------------------------\n')


#Version 2.0
print("\n                                            #Version 2.0 \n")
print("Первоначальная сумма: {} руб".format(source_amt))
print('\n-------------------------------------------------------------------------------\n')
while( time_final != time + 1):
    prib_year = ( (result / 100) * annual_interest)#Прибыль в год
    result = source_amt * pow( (1 + ( annual_interest / 100)), time_final)#Расчет общей суммы накопленной за год
    print("Общая сумма по окончанию {} года равна {} руб, прибыль за год равна {} руб".format(time_final, round(result,2),round(prib_year,2)))
    time_final += 1
prib = (result - source_amt)#общая прибыль
print('\n-------------------------------------------------------------------------------\n')
print("Полученная сумма: {} руб. Общая прибыль: {} руб".format(round(result,2), round(prib,2)))
time_final = 1 #необходимо для нормального перезапуска программы
result = source_amt 
```
